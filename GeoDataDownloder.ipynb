# GeoData downloader
we use this file to Download all the country shapefiles programmatically

The following script downloads the 1:110m Cultural Vectors data, which includes a shapefile for the countries, and then extracts it into the designated directory:

```python
import os
import urllib.request
import zipfile

# Create the necessary directories if not already present
os.makedirs("./Data/GeoData/SHP", exist_ok=True)

# url = "https://www.naturalearthdata.com/http//www.naturalearthdata.com/download/110m/cultural/ne_110m_admin_0_countries.zip"
# url = "https://www.naturalearthdata.com/http//www.naturalearthdata.com/download/10m/cultural/10m_cultural.zip"
# url = "https://www.naturalearthdata.com/http//www.naturalearthdata.com/download/10m/cultural/ne_10m_admin_0_countries_mar.zip"
url = "https://www.naturalearthdata.com/http//www.naturalearthdata.com/download/10m/physical/ne_10m_coastline.zip"
# filename = "./Data/GeoData/SHP/ne_10m_admin_0_countries_mar.zip"
filename = "./Data/GeoData/SHP/ne_10m_coastline.zip"

# Define the user agent
headers = {'User-Agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:15.0) Gecko/20100101 Firefox/15.0.1'}

# Create a request with the defined headers
req = urllib.request.Request(url, headers=headers)

# Use the created request to download the file
with urllib.request.urlopen(req) as response, open(filename, 'wb') as out_file:
    out_file.write(response.read())

# Extract the zip file
with zipfile.ZipFile(filename, 'r') as zip_ref:
    zip_ref.extractall("./Data/GeoData/SHP")

# Remove the zip file if you no longer need it
os.remove(filename)
```
